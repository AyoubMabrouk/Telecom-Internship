```python
# Assuming your dataframe is called 'df'
selected_variables = [
    'OFFRE',
    'revenu_voix',
    'revenu_inter',
    'NB_JOUR_APPEL_TAXE',
    'nb_sms_tot',
    'Duree_onnet_tot',
    'Duree_offnet_tot',
    'NB_ACTIVITE_OUT',
    'revenu_cdr_c',
    'MNT_RECH',
    'NB_RECH',
    'P_revenu_data',
    'NB_SMS_TAXE',
    'NB_FORFAIT',
    'NB_TRANSFERT_OUT',
    'FREQ_ACT_OUT',
    'FREQ_SMS_OUT',
    'FREQ_USSD',
    'VOLUME_SESSION',
    'REVENU_VAS'
]

# Select the desired variables from the dataframe
dataTelecomSelected = dataTelecom[selected_variables].copy()
dataTelecomSelected.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create an instance of LabelEncoder
label_encoder = LabelEncoder()
dataTelecomEncoded=dataTelecomSelected
# Fit and transform the column you want to encode
encoded_offre = label_encoder.fit_transform(dataTelecomEncoded['OFFRE'])

# Replace the original column with the encoded values
dataTelecomEncoded['OFFRE'] = encoded_offre

```
